# 5. Data Visualisation

```python
#Importing necessary libraries

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Histogram

```python
average_nutrition = df.groupby(['Category'], as_index=False)[['Total Fat','Protein', 'Calories', 'Sugars']].mean()
```

```python
average_nutrition
```

```python
#Identifying the density of caloric counts by creating a histogram
plt.figure(figsize =(10,6))
sns.histplot(df['Calories'],bins = 20, kde = True)
plt.xlabel('Calories')
plt.ylabel('Counts')
plt.title('Calorie density of menu items')
plt.show()
```

- Here x-axis represents **Calories** and y-axis represents **Counts**

- The above visualisation show that the most of the values are to the left and highest density lies around 250 calories. Hence, this is Positively skewed as the tail is more pronounced to the left than that of right

```python
#Creating box plot for Distribution pf calorie counts.
plt.figure(figsize = (8,5))
sns.boxplot(x= 'Category',y= 'Calories', data = df , palette = 'magma')
plt.xlabel('CATEGORY')
plt.ylabel('CALORIES')
plt.title('Box plot showing calorie count across Menu Categories')
plt.xticks(rotation = 45)
plt.show()
```

- We have visualised a **Box plot** to get a better understanding of the calorie distribution across various menu categories.Box plot gives a clear view over the distribution of calories and helps in identifying potential Outliers.
- In this plot it is visible that **Smoothies & Shakes** are the menu category with the highest calorie count and **Beverages** are the ones with the lowestcalorie count.
- It is also visible that **Salads** have more calories than **Desserts** which is a bit unusual as salads are considered healthier than desserts.

```python
#Creating a barchart to visualise the distribution of calories across the menu

plt.figure(figsize=(8,5))
sns.barplot(x = 'Category', y ='Calories', data=average_nutrition, palette= 'rainbow')
plt.title('Calories Content Across Menu')
plt.xlabel('Menu Categories')
plt.ylabel('Calories (g)')
plt.xticks(rotation=45)
plt.show()
```

```python
# visualising box plot for distribution of Protein

plt.figure(figsize= (8,5))
sns.boxplot(x= 'Category', y= 'Protein',data = df,  palette= 'rainbow',)
plt.xlabel('Categories')
plt.ylabel('Protein Counts')
plt.title('Boxplot showing Protein distribution across menu')
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x = 'Category', y ='Protein', data=average_nutrition, palette= 'coolwarm')
plt.title('Protein Content Across Menu')
plt.xlabel('Menu Categories')
plt.ylabel('Protein (g)')
plt.xticks(rotation=45)
plt.show()
```

- Both visualisations above represent the protien content across the menu.
- X-axis represents the **Menu categories** and Y-axis represents **Protien in (g)**.
- By reviewing both the plots its is evident that **Chicken&Fish** contain most amount of protein than any other menu category and next comes**Beef&Pork**.
- **Beverages** and **Desserts** contain the least amount of protein.
- By observing the protein distribution it is evident that Mc'Donald's has a decent amount of protien in their foods.

```python
#bar plot showning sugar content in the menu items
plt.figure(figsize=(10,5))
sns.barplot(x = 'Category', y ='Sugars', data=average_nutrition, palette= 'rainbow')
plt.title('Sugar Content Across Menu')
plt.xlabel('Menu Categories')
plt.ylabel('Sugar (g)')
plt.xticks(rotation=45)
plt.show()
```

- In the above plot sugar content in the food categories is represented.
- By the plot it is evident that **Smoothies&Shakes** contain the highest amount of sugars and next comes the **Coffee&Tea** .
